```python
from qadence.draw import display
from qadence import AnsatzConfig, AnsatzType, create_ansatz, Strategy
from qadence import RX, RZ, CNOT

from dataclasses import dataclass, field, fields
```

```python
#============

#ANSATZ

#============
n_qubits = 4
# dictionary to set the ansatz
diz = {'periodic' : False,
       'operations' : [RZ,RZ,RZ,RZ],
       'entangler': CNOT,
       }
ansatz_config = AnsatzConfig(
    depth=2, 
    ansatz_type=AnsatzType.HEA, 
    ansatz_strategy=Strategy.DIGITAL, 
    #not working part
    #=====================
    strategy_args=dict(diz), 
    #=====================
    param_prefix='theta'
    )

ansatz = create_ansatz(register=n_qubits, config=ansatz_config)

print(ansatz_config)

display(ansatz, size="10,10", theme="dark")
```
